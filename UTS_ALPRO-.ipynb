```python
def sequential_search_max(data):
    max_data = data[0]
    for record in data:
        if record["pengunjung"] > max_data["pengunjung"]:
            max_data = record
    return max_data

hasil_sequential = sequential_search_max(data_kunjungan)
print("Hasil Sequential Search:")
print(f"Jam paling ramai: {hasil_sequential['jam']} dengan {hasil_sequential['pengunjung']} pengunjung")
```

```python
def binary_search(data, target):
    left = 0
    right = len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid]["pengunjung"] == target:
            return data[mid]
        elif data[mid]["pengunjung"] < target:
            left = mid + 1
        else:
            right = mid - 1
    return None

# Sorting data dulu berdasarkan pengunjung
data_terurut = sorted(data_kunjungan, key=lambda x: x["pengunjung"])
target_pengunjung = 24

hasil_binary = binary_search(data_terurut, target_pengunjung)
print("\nHasil Binary Search:")
if hasil_binary:
    print(f"Ditemukan: Jam {hasil_binary['jam']} dengan {hasil_binary['pengunjung']} pengunjung")
else:
    print(f"Data dengan {target_pengunjung} pengunjung tidak ditemukan.")
```

*PENCARIAN PENGUNJUNG TERBANYAK DENGAN SEQUENTIAL SEARCH DAN BINARY SEARCH BESERTA WAKTU EKSEKUSI*

```python
def sequential_search_target(data, target):
    steps = 0
    for i, record in enumerate(data):
        steps += 1
        if record["pengunjung"] == target:
            return record, steps
    return None, steps



def binary_search(data, target):
    comparisons = 0
    left = 0
    right = len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        comparisons += 1
        if data[mid]["pengunjung"] == target:
            return data[mid], comparisons
        elif data[mid]["pengunjung"] < target:
            left = mid + 1
        else:
            right = mid - 1
    return None, comparisons

target = 24

# Jalankan Sequential Search
hasil_seq, langkah_seq = sequential_search_target(data_kunjungan, target)
print("Sequential Search:")
if hasil_seq:
    print(f"Jam: {hasil_seq['jam']}, Pengunjung: {hasil_seq['pengunjung']}")
else:
    print("Target tidak ditemukan.")
print(f"Jumlah langkah: {langkah_seq}")
# waktu eksekusi Sequential Search
start_time = time.time()
waktu_eksekusi = time.time() - start_time

print(f"Waktu eksekusi: {waktu_eksekusi:.10f} detik")

# Jalankan Binary Search (dengan data terurut)
data_terurut = sorted(data_kunjungan, key=lambda x: x["pengunjung"])
hasil_bin, langkah_bin = binary_search(data_terurut, target)
print("\nBinary Search:")
if hasil_bin:
    print(f"Jam: {hasil_bin['jam']}, Pengunjung: {hasil_bin['pengunjung']}")
else:
    print("Target tidak ditemukan.")
print(f"Jumlah langkah: {langkah_bin}")
# waktu eksekusi Binary Search
start_time = time.time()
waktu_eksekusi = time.time() - start_time

print(f"Waktu eksekusi: {waktu_eksekusi:.10f} detik")
```

*MMEBUAT GRAFIK HASIL ANALISIS*

```python
# Data dari data_kunjungan
jam = [item['jam'] for item in data_kunjungan]
pengunjung = [item['pengunjung'] for item in data_kunjungan]

# Cari jumlah pengunjung terbanyak dan waktu/jam-nya
max_data = max(data_kunjungan, key=lambda x: x['pengunjung'])
jumlah = max_data['pengunjung']
waktu = max_data['jam']

# Plot grafik
plt.figure(figsize=(12, 6))
plt.plot(jam, pengunjung, marker='o', linestyle='-', color='skyblue', label='Jumlah Pengunjung')
plt.axvline(waktu, color='red', linestyle='--', label=f'Terbanyak: {jumlah} orang ({waktu})')

plt.title("Pengunjung Koperasi per Jam")
plt.xlabel("Jam")
plt.ylabel("Jumlah Pengunjung")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Membuat grafik bar
plt.figure(figsize=(12, 6))
plt.bar(jam, pengunjung, color='skyblue', label='Jumlah Pengunjung')

# Menandai waktu dengan jumlah pengunjung maksimum
plt.bar(waktu, jumlah, color='orange', label=f'Max: {jumlah} pengunjung ({waktu})')

# Menambahkan judul dan label
plt.title('Grafik Bar Jumlah Pengunjung terhadap Waktu', fontsize=14)
plt.xlabel('Waktu', fontsize=12)
plt.ylabel('Jumlah Pengunjung', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.5)
plt.legend()

# Menampilkan grafik
plt.tight_layout()
plt.show()          
```

```python
jam = ['09:10', '09:20', '09:30', '09:40', '09:50','10:00', '10:10', '10:20', '10:30', '10:40', '10:50', '11:00', '11:10', '11:20', '11:30', '11:40', '11:50', '12:00', '12:10', '12:20', '12:30', '12:40', '12:50', '13:00', '13:10', '13:20', '13:30', '13:40', '13:50','14:00']  
pengunjung = [5, 17, 15, 24, 7, 8, 5, 13, 13, 3, 7, 1, 11, 2, 14, 7, 7, 11, 16, 14, 15, 9, 13, 8, 9, 11, 8, 10,7 ,5]
pengunjung_urut = sorted(pengunjung)
jam_urut = sorted(jam)
                         
```

```python
def cari_menit_pertama_lebih_dari_20(pengunjung, jam, batas=20):
    for i in range(len(pengunjung)):
        if pengunjung[i] > batas:
            return jam[i], pengunjung[i]
    return None, None  


waktu, jumlah = cari_menit_pertama_lebih_dari_20(pengunjung, jam)


if waktu:
    print(f"Pertama kali pengunjung lebih 20 terjadi pada menit ke-{waktu}, yaitu {jumlah} orang")
else:
    print("Tidak ada kunjungan yang melebihi 20 pengunjung")
```

```python
def binary_search_pengunjung(pengunjung_urut, jam, batas=20):
    left, right = 0, len(pengunjung_urut) - 1
    result = None

    while left <= right:
        mid = (left + right) // 2
        if pengunjung_urut[mid] > batas:
            result = (jam[mid], pengunjung_urut[mid])
            right = mid - 1  # Cari di bagian kiri untuk menemukan waktu pertama
        else:
            left = mid + 1

    return result


# Menggunakan binary search
binary_result = binary_search_pengunjung(pengunjung_urut, jam)

if binary_result:
    binary_waktu, binary_jumlah = binary_result
    print(f"Pertama kali pengunjung lebih dari 20 terjadi pada jam {binary_waktu}, yaitu {binary_jumlah} orang")
else:
    print("Tidak ada kunjungan yang melebihi 20 pengunjung")
```

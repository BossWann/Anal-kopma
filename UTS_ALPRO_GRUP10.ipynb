### *PENCARIAN PENGUNJUNG TERBANYAK DENGAN SEQUENTIAL SEARCH DAN BINARY SEARCH BESERTA WAKTU EKSEKUSI*

```python
def sequential_search_target(data, target):
    steps = 0
    for i, record in enumerate(data):
        steps += 1
        if record["pengunjung"] == target:
            return record, steps
    return None, steps



def binary_search(data, target):
    comparisons = 0
    left = 0
    right = len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        comparisons += 1
        if data[mid]["pengunjung"] == target:
            return data[mid], comparisons
        elif data[mid]["pengunjung"] < target:
            left = mid + 1
        else:
            right = mid - 1
    return None, comparisons

target = 24

# Jalankan Sequential Search
hasil_seq, langkah_seq = sequential_search_target(data_kunjungan, target)
print("Sequential Search:")
if hasil_seq:
    print(f"Jam: {hasil_seq['jam']}, Pengunjung: {hasil_seq['pengunjung']}")
else:
    print("Target tidak ditemukan.")
print(f"Jumlah langkah: {langkah_seq}")
# waktu eksekusi Sequential Search
start_time = time.time()
waktu_eksekusi = time.time() - start_time

print(f"Waktu eksekusi: {waktu_eksekusi:.10f} detik")

# Jalankan Binary Search (dengan data terurut)
data_terurut = sorted(data_kunjungan, key=lambda x: x["pengunjung"])
hasil_bin, langkah_bin = binary_search(data_terurut, target)
print("\nBinary Search:")
if hasil_bin:
    print(f"Jam: {hasil_bin['jam']}, Pengunjung: {hasil_bin['pengunjung']}")
else:
    print("Target tidak ditemukan.")
print(f"Jumlah langkah: {langkah_bin}")
# waktu eksekusi Binary Search
start_time = time.time()
waktu_eksekusi = time.time() - start_time

print(f"Waktu eksekusi: {waktu_eksekusi:.10f} detik")
```

### *ANALISIS PERIODE WAKTU DENGAN PENGUNJUNG LEBIH DARI 10 DENGAN BINARY SEARCH DAN SEQUENTIAL SEARCH*

```python
def cari_periode_pengunjung_lebih_dari(data, batas):
    hasil = []
    for record in data:
        if record["pengunjung"] > batas:
            hasil.append(record)
    return hasil

# Analisis periode waktu dengan pengunjung lebih dari 10
batas_pengunjung = 10
periode_lebih_dari_10 = cari_periode_pengunjung_lebih_dari(data_kunjungan, batas_pengunjung)

print("Periode waktu dengan pengunjung lebih dari 10:")
for periode in periode_lebih_dari_10:
    print(f"Jam: {periode['jam']}, Pengunjung: {periode['pengunjung']}")

    # Hitung jumlah langkah pengerjaan
    jumlah_langkah = len(data_kunjungan)
    print(f"\nJumlah langkah pengerjaan: {jumlah_langkah}")
```

Penjelasan Jumlah Langkah Pengerjaan = 30

Jumlah langkah pengerjaan adalah 30 karena:

1. Fungsi `cari_periode_pengunjung_lebih_dari()` menggunakan sequential search yang memeriksa setiap elemen dalam data
2. Jumlah data dalam `data_kunjungan` adalah 30 record
3. Sequential search harus memeriksa seluruh data untuk memastikan semua pengunjung yang lebih dari batas ditemukan
4. Tidak ada cara untuk menghentikan pencarian lebih awal karena kita perlu menemukan semua periode yang memenuhi kriteria
5. Oleh karena itu, jumlah langkah akan selalu sama dengan jumlah total data = 30

Berbeda dengan binary search yang bisa lebih efisien untuk mencari satu nilai spesifik, sequential search diperlukan di sini untuk menemukan semua nilai yang memenuhi kriteria.

```python
def binary_search_greater_than(data, batas):
    hasil = []
    left = 0
    right = len(data) - 1

    # Cari elemen pertama yang lebih besar dari batas
    while left <= right:
        mid = (left + right) // 2
        if data[mid]["pengunjung"] > batas:
            right = mid - 1
        else:
            left = mid + 1

    # Tambahkan semua elemen yang lebih besar dari batas ke hasil
    for i in range(left, len(data)):
        if data[i]["pengunjung"] > batas:
            hasil.append(data[i])
        else:
            break

    return hasil

# Analisis periode waktu dengan pengunjung lebih dari 10 menggunakan binary search
periode_lebih_dari_10_bin = binary_search_greater_than(data_terurut, batas_pengunjung)

print("Periode waktu dengan pengunjung lebih dari 10 (Binary Search):")
for periode in periode_lebih_dari_10_bin:
    print(f"Jam: {periode['jam']}, Pengunjung: {periode['pengunjung']}")

    # Hitung jumlah langkah pengerjaan untuk binary search
    jumlah_langkah_binary = len(data_terurut) // 2
    print(f"\nJumlah langkah pengerjaan binary search: {jumlah_langkah_binary}")
```

Penjelasan kenapa jumlah langkah binary search adalah 15

Jumlah langkah pengerjaan binary search adalah 15 karena:

1. Binary search membagi data menjadi 2 bagian pada setiap iterasi
2. Dengan 30 data, maka:
    `- Iterasi 1: 30/2 = 15 data
    - Iterasi 2: 15/2  8 data
    - Iterasi 3: 8/2 = 4 data
    - Iterasi 4: 4/2 = 2 data
    - Iterasi 5: 2/2 = 1 data
    
3. Rumus untuk menghitung jumlah langkah maksimal binary search:
    jumlah_langkah = len(data) // 2 
    jumlah_langkah = 30 // 2 = 15
